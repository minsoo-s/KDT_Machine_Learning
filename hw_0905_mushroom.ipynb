##### [2] input, target 분리


```python
# 새로운 데이터프레임 생성
df_num = pd.DataFrame(columns= df.columns)
df_num = df_num.T
```

```python
# 알파벳을 숫자로 치환
#arr_df = np.array(df)
idx = 'abcdefghijklmnopqrstuvwxyz'

i=0
for row in df.values:
    row_list = []

    for alphabet in row:
        row_list.append(idx.find(alphabet))
    df_num[i]=row_list
    i+=1
```

```python
df_num = df_num.T
df_num
```

```python
target = df_num['p']
target

input = df_num.iloc[:,1:]
input
```

```python
print(input.shape)
print(target.shape)

input = np.array(input)
target = np.array(target)

print(input.shape)
print(target.shape)
```

```python
# 훈련, 테스트, 검증 -> 6 : 2 : 2
from sklearn.model_selection import train_test_split

# 훈련, 테스트용 분리
train_input, test_input, train_target, test_target, = train_test_split(input, target, test_size = 0.2, random_state = 42)

# 훈련, 검증용 분리
train_input, val_input, train_target, val_target, = train_test_split(train_input, train_target, test_size = 0.25, random_state = 42)
```

##### [3] 모델 생성, 훈련, 정확도 확인

```python
train_input
```

```python
from sklearn.utils import all_estimators

# sklearn의 모든 필터타입 분류기 호출 -> list(이름,객체) 
model_all = all_estimators(type_filter='classifier')

result = []
for name, model in model_all:
    try:
        md = model()
        md.fit(train_input,train_target)
        score = md.score(test_input, test_target)
        result.append((name, score))
    except:
        pass
```

```python
result
```

##### [3] 모델 생성, 훈련, 정확도 확인

```python
train_target
```

```python
from sklearn.ensemble import RandomForestClassifier

model_RF = RandomForestClassifier()
model_RF.fit(train_input,train_target)

print('테스트 score:', model_RF.score(test_input,test_target))
print('검증용 score:', model_RF.score(val_input,val_target))
```

```python
# 교차검증 함수
from sklearn.model_selection import cross_validate

result_RF = cross_validate(model_RF,input,target,return_train_score=True)
```

```python
print('[ 랜덤포레스트 점수 ]')
print('LR_train score:',result_RF['train_score'].mean())
print('LR_test score:',result_RF['test_score'].mean())
```
